```python
adj = {
    '0000':[('1111', 4), ('1155', 12), ('5511', 12)],
    '1111':[('0000', 4), ('1155', 8), ('5511', 8)],
    '1155':[('1111', 8), ('0000', 12), ('5511', 16)],
    '5511':[('1111', 8), ('1155', 16), ('0000', 12)]
}

def prim(adj):
    Vt = ['0000']
    Xt = []
    for i in range(len(adj) - 1):
        min_e = None
        min_v = None
        for v in Vt:
            for vecino in adj[v]:
                if vecino[0] in Vt: continue
                if min_e == None or min_e[1] > vecino[1]:
                    min_e = vecino
                    min_v = vecino[0]
        
        Xt.append(min_e)
        Vt.append(min_v)
    return Vt, Xt
prim(adj)
```

```python
import heapq

def prim(adj, start_vertex):
    Vt = set([start_vertex])
    Xt = []
    edges = []
    for vecino in adj[start_vertex]:
        heapq.heappush(edges, (vecino[1], start_vertex, vecino[0]))

    while len(Vt) < len(adj):
        weight, u, v = heapq.heappop(edges)

        if v not in Vt:
            Vt.add(v)
            Xt.append((u, v, weight))
            for vecino in adj[v]:
                if vecino[0] not in Vt:
                    heapq.heappush(edges, (vecino[1], v, vecino[0]))

    return list(Vt), Xt

def dist(v, w):
    acc = 0
    for i in range(len(v)):
        print(v, w)
        c1, c2 = int(v[i]), int(w[i])
        diff = abs(c1 - c2)
        if diff <= 5:
            acc += diff
        else:
            acc += 10 - diff
    return acc

cases = int(input())
for k in range(cases):
    i = input()
    parsed = ['0000'] + i.split(' ')[1:]
    adj = {v: [] for v in parsed}

    for j in range(len(parsed)):
        v = parsed[j]
        for k in range(len(parsed)):
            w = parsed[k]
            if v == w: continue
            d = dist(v, w)
            adj[v].append((w, d))
            adj[w].append((v, d))

    start_vertex = '0000'
    Vt, Xt = prim(adj, start_vertex)
    min_num = sum([e[2] for e in Xt])
    print(min_num)
```
