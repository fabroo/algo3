```python
def minimacantidadPD(v, i, last_index, memo = None):
    if memo == None:
        memo = [[None] * (n + 1) for _ in range (n + 1)]
        
    if i == -1:
        return 0
    
    res = 0
    if memo[i][last_index] != None:
        res = memo[i][last_index]
    
    if (v[last_index] == 0 and last_index == n - 1) or v[last_index] % v[i] == 0:
        res = min(minimacantidadPD(v, i - 1, i, memo), 1 + minimacantidadPD(v, i - 1, last_index, memo)) 
    else:
        res = 1 + minimacantidadPD(v, i - 1, last_index, memo)
    
    memo[i][last_index] = res
    return memo[i][last_index]

v = [2, 4]
n = len(v)

result = minimacantidadPD(v, n - 1, n - 1)
print("Minimum operations required:", result)
```
