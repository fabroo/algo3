```python
def isMagic(square):
    n = len(square)
    target_sum = sum(square[0])
    
    for row in square:
        if sum(row) != target_sum:
            return False
    for j in range(n):
        col_sum = sum(square[i][j] for i in range(n))
        if col_sum != target_sum:
            return False
    main_diag_sum = sum(square[i][i] for i in range(n))
    anti_diag_sum = sum(square[i][n - i - 1] for i in range(n))
    if main_diag_sum != target_sum or anti_diag_sum != target_sum:
        return False
    
    return True 

def isValid(T, k):
    for j in range(len(T)):
        for i in range(len(T)):
            if T[i][j] == k:
                return False
    return True

def isValidPodas(T, k, m_number):
    n = len(T)
    
    # Check for repeated values in columns
    for j in range(n):
        for i in range(n):
            if T[i][j] == k:
                return False
    
    # Check if any row sum exceeds m_number
    for row in T:
        if sum(row) > m_number:
            return False
    
    # Check if any column sum exceeds m_number
    for j in range(n):
        column_sum = 0
        for i in range(n):
            column_sum += T[i][j]
        if column_sum > m_number:
            return False
    
    # Check diagonal sums
    main_diag_sum = sum(T[i][i] for i in range(n))
    anti_diag_sum = sum(T[i][n - 1 - i] for i in range(n))
    if main_diag_sum > m_number or anti_diag_sum > m_number:
        return False
    
    return True
```

```python
def BT(T, i, j, k):
    n = len(T)
    if j == n:
        return BT(T, i + 1, 0, 1)
    
    elif i == n:
        if isMagic(T):
            solutions.append([row[:] for row in T])  # Deep copy of the matrix
            return 1
        return 0
    
    else:
        count = 0
        for k in range(1, n*n + 1):
            if isValid(T, k):                
                T[i][j] = k
                count += BT(T, i, j + 1, k + 1)
                T[i][j] = 0  

        return count

def BT_podas(T, i, j, k):
    n = len(T)
    magic_number = (n ** 3 + n) /2
    if j == n:
        return BT(T, i + 1, 0, 1)
    
    elif i == n:
        if isMagic(T):
            solutions.append([row[:] for row in T])  # Deep copy of the matrix
            return 1
        return 0
    
    else:
        count = 0
        for k in range(1, n*n + 1):
            if isValidPodas(T, k, magic_number):                
                T[i][j] = k
                count += BT(T, i, j + 1, k + 1)
                T[i][j] = 0  

        return count
```
